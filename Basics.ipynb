# Predicting Car Prices
In this project, we will use a dataset containing the price of the cars and other attributes of cars.Our goal is to find apropriate arrtibutes to develop a model for us to predict car prices.

## Data Introduction
First, we will introduct and explore the dataset to find its features and issues.

```python
import pandas as pd
import numpy as np

cars = pd.read_csv('imports-85.data')
cars.head()
```

We notice that the colomn names are not in the dataset, so we will read the documentation and add correspondong column names for it.

```python
cols = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 
        'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type', 
        'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']

cars = pd.read_csv('imports-85.data', names=cols)
cars.head()
```

```python
# FInd the type for each column
cars.info()
```

```python
# Find if there are null values
cars.isnull().sum()
```

Literally, there are no missing values, but as we inspect the first few lines of the dataset, we can find that column `normalized-losses` using symbol `?` to represent missing values.

```python
cars['normalized-losses'].value_counts()[:5]
```

## Data Cleaning
After exploring the dataset, there are few steps to clean the dataset:
* replace all `?` with `numpy.nan` to help us identify null values.
* covert columns originally had `?` to numeric type.
* Make a rule on resolving missing values.

```python
# replace all ? with NAN
cars = cars.replace('?', np.nan)

# Select only columns with numeric values
continuous_values_cols = ['normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-size', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
num_cars = cars[continuous_values_cols].copy()
```

```python
# Convert all columns to numeric type
num_cars = num_cars.astype(float)
num_cars.head()
```

```python
# Find all missing values
num_cars.isnull().sum()
```

As our goal is to predict car price, and missing values in `price` are too little to influense the population. As a result, we will first remove rows whose `price` is missing.

```python
num_cars = num_cars.dropna(subset=['price'])
num_cars.isnull().sum()
```

For the remaining rows, we will use the corresponding **mean values** to replace the missing values.

```python
num_cars = num_cars.fillna(num_cars.mean())
num_cars.isnull().sum()
```

### Normalization
Now we have remove all missing values and select only the numeric columns, we will continue to **normalize each columns**
> Different column have different range, so normalization is essntial.

In this project, we will use **Min-Max Normalization** to make all values **range from 0 to 1**.

```python
# Apparently, we don't want to normalize the price column!
price_col = num_cars['price']

num_cars = (num_cars - num_cars.min()) / (num_cars.max() - num_cars.min())
num_cars['price'] = price_col
num_cars.head()
```

## Model Building
All columns in our dataset are now ready to go. We will start building our model using **univariate k-nearest neighbors** first to better understanding features of data.

### Univariate Model
We will write a function that training with our data using only one columns, and it should return the **RMSE** for corresponding model's prediction result.

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
import math

# The function takes in three parameters
# train_cols: column used for training
# target_col: columns to predict
# df the whole dataset
def knn_train_test(train_cols, target_col, df):
    # Make randomization reproducible
    np.random.seed(1)
    
    # Suffle the dataset
    random_index = np.random.permutation(df.index)
    new_df = df.reindex(random_index)
    
    # Split into training 75% & test set 25%
    split_point = math.ceil(new_df.shape[0] * 0.75)
    train_set = new_df.iloc[:split_point]
    test_set = new_df.iloc[split_point:]
    
    # Initialize model object with default params
    knn = KNeighborsRegressor()
    
    # Train and predict
    knn.fit(train_set[train_cols], train_set[target_col])
    prediction = knn.predict(test_set[train_cols])
    
    # Calculate RMSE
    rmse = round(mean_squared_error(test_set[target_col], prediction) ** (1 / 2), 4)
    return rmse
```

Let's test it with each column in the dataset.

```python
all_cols = list(num_cars.columns)
all_cols.remove('price')
univariate_rmse  = {}
for col in all_cols:
    rmse = knn_train_test([col], 'price', num_cars)
    univariate_rmse[col] = rmse
    
# Display the result
for k, v in univariate_rmse.items():
    print(k + ": " + str(v))

# Present the mean RMSE
mean_rmse = sum(univariate_rmse.values()) / len(univariate_rmse)
print("\nMean RMSE: " + str(mean_rmse))
```

Instead of using the **default k(5)**, we will add an parameter to our function to use different k values for out model.

```python
# Modified function
def knn_train_test_k(train_cols, target_col, df, k):
    # Make randomization reproducible
    np.random.seed(1)
    
    # Suffle the dataset
    random_index = np.random.permutation(df.index)
    new_df = df.reindex(random_index)
    
    # Split into training 75% & test set 25%
    split_point = math.ceil(new_df.shape[0] * 0.75)
    train_set = new_df.iloc[:split_point]
    test_set = new_df.iloc[split_point:]
    
    # Initialize model object with default params
    knn = KNeighborsRegressor(n_neighbors=k)
    
    # Train and predict
    knn.fit(train_set[train_cols], train_set[target_col])
    prediction = knn.predict(test_set[train_cols])
    
    # Calculate RMSE
    rmse = round(mean_squared_error(test_set[target_col], prediction) ** (1 / 2), 4)
    return rmse
```

Now, we will test our function, and visualize the results to get a better picture.

```python
# k values for test
k_vals = [k for k in range(1, 20, 2)]
# store the result
col_rmse = {}

# for every column, test every k value
for col in all_cols:
    k_rmse = {}
    for k in k_vals:
        rmse = knn_train_test_k([col], 'price', num_cars, k)
        k_rmse[k] = rmse
    col_rmse[col] = k_rmse

# Visualize the results
import matplotlib.pyplot as plt
import seaborn as sns
import collections
%matplotlib inline

plt.figure(figsize=(15, 25))
i = 1
for col, k_map in col_rmse.items():
    plt.subplot(7, 2, i)
    ok = collections.OrderedDict(sorted(k_map.items()))
    plt.scatter(list(ok.keys()), list(ok.values()))
    plt.title(col)
    plt.ylim((2500, 10500))
    plt.axhline(min(list(k_map.values())), color='red')
    i += 1

plt.show()
```

It seems that when k value is over **10**, the RMSE will always generally higher. Thus we will drop k that are over 10, and plot a line graph to view the difference.

```python
# k values for test
k_vals = [1, 3, 5, 7, 9]

# store the result
col_rmse = {}

# for every column, test every k value
for col in all_cols:
    k_rmse = {}
    for k in k_vals:
        rmse = knn_train_test_k([col], 'price', num_cars, k)
        k_rmse[k] = rmse
    col_rmse[col] = k_rmse

plt.figure(figsize=(8, 5))
for col, k_map in col_rmse.items():
    ok = collections.OrderedDict(sorted(k_map.items()))
    plt.plot(list(ok.keys()), list(ok.values()), label=col)
    plt.ylim((2500, 10500))

plt.legend(loc=(1.01, 0.5), ncol=2)
plt.show()
```

It is noticable that the plot is divided into two parts. What we want is columns that have **lower RMSE**, so we will **drop** columns that have their **lowest RMSE over 6000**.

```python
# Store columns to be deleted
error_cols = []

for col, k_map in col_rmse.items():
    rmse_vals = list(k_map.values())
    if min(rmse_vals) > 6000:
        error_cols.append(col)
    
# Inspect error columns
error_cols
```

```python
# Remove there col
all_cols = [ele for ele in all_cols if ele not in error_cols]

# Draw a clearer line plot
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'purple', 'black', 'aqua']
color_index = 0

plt.figure(figsize=(8, 5))
for col, k_map in col_rmse.items():
    if col in error_cols:
        continue;
    ok = collections.OrderedDict(sorted(k_map.items()))
    plt.plot(list(ok.keys()), list(ok.values()), label=col, color=colors[color_index])
    plt.ylim((2500, 6500))
    color_index += 1

plt.legend(loc=(1.01, 0.5), ncol=2)
plt.show()
```

### MultiVariate Model
Since we have found several columns that performed well in previous model, we can now move on to model that takes in multiple columns for training. Firstly, we will determine what k value we should use by finding **minimum mean RMSE** values.

```python
# Store mean RMSE of each k
mean_rmse = {}

for col, k_map in col_rmse.items():
    if col in error_cols:
        continue
    for k, v in k_map.items():
        if k in mean_rmse:
            mean_rmse[k] += v
        else:
            mean_rmse[k] = v

mean_rmse
```

We can see k value of 1 and 3 have smaller RMSE. However, too small a k value will cause large noise in the outcome. Therefore, **balancing noise and RMSE**, we will eventually use **k=5** to build our model

```python
# Because k=5, we will use knn_train_test() at first
# Find best features
best_col_map = {}
for col, k_map in col_rmse.items():
    if col in error_cols:
        continue
    for k, v in k_map.items():
        if k == 5:
            best_col_map[col] = v
            break

best_col_map
```

```python
best_two_features = ['engine-size', 'highway-mpg']
best_three_features = ['engine-size', 'highway-mpg', 'city-mpg']
best_four_features = ['engine-size', 'highway-mpg', 'city-mpg', 'width']
best_five_features = ['engine-size', 'highway-mpg', 'city-mpg', 'width', 'curb-weight']
best_six_features = ['engine-size', 'highway-mpg', 'city-mpg', 'width', 'curb-weight', 'wheel-base']
best_seven_features = ['engine-size', 'highway-mpg', 'city-mpg', 'width', 'curb-weight', 'wheel-base', 'horsepower']

multivariate_rmse = {}
index = 2
for multi_features in [best_two_features, best_three_features, best_four_features, best_five_features, best_six_features, best_seven_features, all_cols]:
    rmse = knn_train_test(multi_features, 'price', num_cars)
    multivariate_rmse[index] = rmse
    index += 1

multivariate_rmse
```

## Hyperparameter Optimization 
Finally, we will do **hyperparameter optimization** to evaluate our model. We will set k from 1 to 25. Will drop two worst model as well, which is two features and three features.

```python
# Modify the function, return amp for k and RMSE
def knn_train_test_k_map(train_cols, target_col, df):
    # Make randomization reproducible
    np.random.seed(1)
    
    # Suffle the dataset
    random_index = np.random.permutation(df.index)
    new_df = df.reindex(random_index)
    
    # Split into training 75% & test set 25%
    split_point = math.ceil(new_df.shape[0] * 0.75)
    train_set = new_df.iloc[:split_point]
    test_set = new_df.iloc[split_point:]
    
    k_map = {}
    for k in range(1, 26):
         # Initialize model object with default params
        knn = KNeighborsRegressor(n_neighbors=k)

        # Train and predict
        knn.fit(train_set[train_cols], train_set[target_col])
        prediction = knn.predict(test_set[train_cols])

        # Calculate RMSE
        rmse = round(mean_squared_error(test_set[target_col], prediction) ** (1 / 2), 4)
        k_map[k] = rmse
        
    return k_map
```

```python
hyper_optim_map = {}

# Iterate every model to do hyperparameter optimization
index = 4
for multi_features in [best_four_features, best_five_features, best_six_features, best_seven_features, all_cols]:
    rmse_map = knn_train_test_k_map(multi_features, 'price', num_cars)
    hyper_optim_map['best {} features'.format(index)] = rmse_map
    index += 1

hyper_optim_map
```

```python
# Draw conprehensive plots to show the result
plt.figure(figsize=(10, 12))
for i, (key, value) in zip(range(1, 6), hyper_optim_map.items()):
    plt.subplot(3, 2, i)
    plt.plot(list(value.keys()), list(value.values()))
    plt.ylim((2000, 4100))
    plt.title(key)
    
plt.show()

for key, value in hyper_optim_map.items():
    plt.plot(list(value.keys()), list(value.values()), label=key)
    plt.ylim((2000, 4100))

plt.legend(loc=(1.01, 0.5), ncol=2)
plt.show()
```
